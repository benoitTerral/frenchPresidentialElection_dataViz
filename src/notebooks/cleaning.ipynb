I've got very strange results for the first round but they are in line with the [official french data](https://www.archives-resultats-elections.interieur.gouv.fr/resultats/presidentielle-2022/032/002/002078.php).
Moving to the largest city of these cities, All the results are set to [Nuls](https://www.archives-resultats-elections.interieur.gouv.fr/resultats/presidentielle-2022/094/02A/02A065.php). Digging further, the invalidation is due to the french Conseil constitutionnel institution invalidating due to irregulaties. In [this article](https://www.conseil-constitutionnel.fr/decision/2022/2022195PDR.htm), all these unexpected results are explained.

```python
# do I want to check the result ?
candidate_vote = dataframes['2022'].filter(regex='^voix(?!.*ratio)(?!.*T2)')
sum_results = candidate_vote.sum(axis=1)
sum_results
```

### Loop on all the datasets

First, the analysis is at the level of the `departement`, let's make sure this data is properly filled in all the dataset

```python
for year, df in dataframes.items():
	print(year)
	print(f"missing dep code: {df['dep'].isna().sum()}")
	print(f"missing dep name: {df['nomdep'].isna().sum()}")
	print(f"total nb of lines: {df.shape[0]}")
```

Many elections years before 1995 have some cities not attached to a departement, let's focus on dates after that one.

```python
year_to_remove = ['1848', '1965', '1969', '1974', '1981', '1988']
dataframes = {key: value for key, value in dataframes.items() if key not in year_to_remove}
```

```python
for year, df in dataframes.items():
	print(year)
	print(f"missing dep code: {df['dep'].isna().sum()}")
	print(f"missing dep name: {df['nomdep'].isna().sum()}")
	print(f"total nb of lines: {df.shape[0]}")
```
