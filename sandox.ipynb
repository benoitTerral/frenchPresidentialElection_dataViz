### Import

```python
import pandas as pd
import wget
from zipfile import ZipFile
import os
```

```python
url = "https://conflit-politique-data.ams3.cdn.digitaloceanspaces.com/zip/pres2022_csv.zip"
try:
	zip_file = wget.download(url)
except Exception as e:
	raise e
src_filename = 'pres2022_csv/pres2022comm.csv'
dest_filename = 'data.csv'
with ZipFile(zip_file, 'r') as zipObject:
	zipObject.getinfo(src_filename).filename = dest_filename
	zipObject.extract('pres2022_csv/pres2022comm.csv')
zipObject.close()
# delete the zip 
if (os.path.exists(zip_file)):
	os.remove(zip_file)
try:
	df = pd.read_csv('data.csv', dtype={'dep': 'string', 'codecommune': 'string'})
except pd.errors.ParserWarning as e:
	print(e)
```

```python
print(f'Nombre de commune: {"{:,}".format(df.shape[0])}')
print(f'Nombre d\'inscrits : {"{:,}".format(df["inscrits"].sum())}')
```
